### 3)  formulating targeted public health policies and interventions to combat tuberculosis (TB) and HIV co-infection in different regions or countries/territories?

#### 3.1) AVG TB & HIV prevalance

```python
avg_tb= df.groupby('Region')['Estimated prevalence of TB (all forms) per 100 000 population'].mean()
avg_hiv= df.groupby('Region')['Estimated incidence of TB cases who are HIV-positive per 100 000 population'].mean()
print(avg_hiv)
print(avg_tb)
```

#### 3.2) Visualize the average prevalence of TB and HIV-positive cases by region

```python
sns.barplot(x=avg_tb.index, y=avg_tb.values, color='skyblue', label='Average TB Prevalence')
sns.barplot(x=avg_hiv.index, y=avg_hiv.values, color='salmon', label='Average HIV Prevalence')
plt.xlabel('Region')
plt.ylabel('Average Prevalence per 100,000 population')
plt.title('Average Prevalence of TB and HIV by Region')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 3.3) Analyze correlations between TB and HIV prevalence

```python
df = df.rename(columns={'Estimated prevalence of TB (all forms) per 100 000 population': 'TB Prevalence'})
df = df.rename(columns={'Estimated incidence of TB cases who are HIV-positive per 100 000 population': 'HIV Prevalence'})
```

```python
correlation = df[['TB Prevalence', 'HIV Prevalence']].corr()
print(correlation)
```

### Visualize correlation using a heatma

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between TB and HIV Prevalence')
plt.tight_layout()
plt.show()
```

```python
high_tb_hiv = df[(df['TB Prevalence'] > 500) & 
                 (df['HIV Prevalence'] > 50)]

print("Countries/Territories with High TB and HIV Prevalence:")
print(high_tb_hiv[['Country or territory name', 'Region', 'TB Prevalence', 
                    'HIV Prevalence']])
```
